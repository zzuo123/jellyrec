```python
from scipy import spatial
a = [1, 2]
b = [2, 4]
c = [2.5, 4]
d = [4.5, 5]
```

```python
import matplotlib.pyplot as plt
def plot(a, b, c, d):
    # List of all points
    points = [a, b, c, d]
    labels = ['a', 'b', 'c', 'd']

    # Extracting x and y coordinates
    x_coords = [point[0] for point in points]
    y_coords = [point[1] for point in points]

    # Plotting the points
    plt.scatter(x_coords, y_coords)

    # Plotting lines from the origin to each point
    for x, y, label in zip(x_coords, y_coords, labels):
        plt.plot([0, x], [0, y], label=f'Line to {label}')

    # Adding labels and title
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Lines from Origin to Points')

    # Labeling each point with letters
    for i, label in enumerate(labels):
        plt.annotate(label, (x_coords[i], y_coords[i]), textcoords="offset points", xytext=(0,10), ha='center')

    # Centering the origin
    ax = plt.gca()
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    
    # Adding a legend
    plt.legend()

    # Display the plot
    plt.show()
```

```python
def standardize(lst):
    print("ORG:", lst)
    avg = sum(lst)/len(lst)
    print("AVG:", avg)
    res = list(map(lambda x:x-avg, lst))
    print("NRM:", res)
    return res
```

```python
a = standardize(a)
b = standardize(b)
c = standardize(c)
d = standardize(d)
```

```python
plot(a,b,c,d)
```
